```python
import pandas as pd
import re
from matplotlib import pyplot as plt
from nltk.tokenize import TweetTokenizer
```

```python
#Q1

tokenizer = TweetTokenizer(r'\w+')

data = " ".join(pd.read_csv("E:/tweets-dataset.csv")['Sentence'])

data = re.sub("http\S+"," ",data)

string = tokenizer.tokenize(''.join([i for i in data.lower()]))

data = re.sub("[^a-zA-Z#]"," ",data)

print('tokens = '+ str(len(string)) +'\ttypes ='+str(len(set(string))) +'\tTTP = '+str(round(len(set(string))/len(string),6)))
```

```python
#Q3
set_string = set(string)

V = []
types = set()
for i in range (len(string)):
    types.add(string[i])
    if i %10000 == 0:
        V.append(len(types))

N = []
for i in range (len(V)):
    N.append(10000*i)

k = 50
B = 0.52
theo_V = []
for i in range (len(V)):
    theo_V.append(k*(N[i]**B))
```

```python
plt.plot(N[:], V[:], '-r',label='Experimental Curve')
plt.plot(N[:],theo_V[:],'-b',label = 'Theoritical Curve')
plt.legend(loc='upper left')
plt.title('Heap\'s Law')
plt.xlabel('N')
plt.ylabel('|V|')
# plt.text(225000, 25000, 'K = ' + str(round(k,2)) + '\nBeta = ' + str(round(beta,2)), dict(size=18))
plt.show()
```
